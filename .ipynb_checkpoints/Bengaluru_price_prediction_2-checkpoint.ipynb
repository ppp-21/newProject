##### Data type is object, some of the values are not correct

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

##### We then us "~" to neglect the rows with floating values

```python
df5[~df5['total_sqft'].apply(is_float)].head(10)
```

##### We will take mean of the object values and some values have different unit we will remove thsoe values

```python
def convert_sqft_to_num(x):
    value = x.split('-')
    if len(value) == 2:
        return (float(value[0])+float(value[1]))/2
    try:
        return float(x)
    except:
        return None 
```

```python
df6 = df5.copy()
df6['total_sqft'] = df6['total_sqft'].apply(convert_sqft_to_num)
df6
```

# Feature Engineering

## 1. Linearity

```python
df7.corr()[['price']][0:5]
```

```python
plt.figure(figsize=(8,2))
sns.heatmap(df7.corr()[['price']][0:5].T,annot=True)
plt.savefig('Correlation.png')
```

#### We can drop balcony column as it has low linearity with price

```python
df8 = df7.copy()
```

```python
df8.drop(['balcony'],axis=1,inplace=True)
```

## 2. No multicolinearity

```python
rows,columns = df8.shape
print('Rows:',rows)
print('Columns:',columns)
```

```python
for i in range(6):
    vif = variance_inflation_factor(df8.to_numpy(),i)
    print(f"VIF for {df.columns[i]}:",vif)
```

## Model Training

```python
x= df8.drop('price',axis=1)
y= df8['price']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=46)
x_train
```

```python
linear_reg = LinearRegression()
linear_reg.fit(x_train,y_train)
```

## Training Data

```python
y_pred_train = linear_reg.predict(x_train)
mse = mean_squared_error(y_train,y_pred_train)
print('MSE for Training:',mse)

mae = mean_absolute_error(y_train,y_pred_train)
print('MAE for Training:',mae)

r2 = r2_score(y_train,y_pred_train)
print('R2 Score for Training:',r2)
```

## Testing Data

```python
y_pred = linear_reg.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print('MSE for Testing:',mse)

mae = mean_absolute_error(y_test,y_pred)
print('MAE for Testing:',mae)

r2 = r2_score(y_test,y_pred)
print('R2 Score for Testing:',r2)
```
